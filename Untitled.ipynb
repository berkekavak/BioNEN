```python
import requests
import re

# Set the protein name
protein_name = "Retinoblastoma-associated protein"
protein_name = "human papillomavirus"
protein_name = "E1A"
protein_name = protein_name.replace(' ','%20')

print(protein_name)
# Set the UniProt search URL
search_url = "https://rest.uniprot.org/uniprotkb/search?query=(reviewed:true)%20AND%20(protein_name:{0})&fields=accession".format(protein_name)

# Send the HTTP GET request to search for the protein
response = requests.get(search_url)
requests.get(search_url).text

# Extract the UniProt ID from the response using regular expressions or HTML parsing library like BeautifulSoup

# If using regular expressions

uniprot_id = re.findall('data-isoform-id="([A-Z0-9]{6})"', response.text)

# If using BeautifulSoup
# from bs4 import BeautifulSoup
# soup = BeautifulSoup(response.text, 'html.parser')
# uniprot_id = [tag['data-isoform-id'] for tag in soup.find_all('a', {'data-isoform-id': True})]

# Check if the UniProt ID was found
if uniprot_id:
    print("UniProt ID for {0}: {1}".format(protein_name, uniprot_id[0]))
else:
    print("UniProt ID not found for {0}".format(protein_name))
```
