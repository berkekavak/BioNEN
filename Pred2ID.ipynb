```python
species_dct = {}
gene_dct = {}
disease_dct = {}
chemical_dct = {}
variant_dct = {}

with open('data/DrugProt_Train.PubTator') as f:
    lines = [line.rstrip('\n') for line in f]
    for line in lines:
        if len(line.split('\t'))>1:
            if(line.split('\t')[-2]=='Species'):
                species_dct[line.split()[3]] = line.split()[-1]                  
            if(line.split('\t')[-2]=='Gene'):
                gene_dct[line.split()[3]] = line.split()[-1]
            if(line.split('\t')[-2]=='Disease'):
                disease_dct[line.split()[3]] = line.split()[-1]
            if(line.split('\t')[-2]=='Chemical'):
                chemical_dct[line.split()[3]] = line.split()[-1]
            if(line.split('\t')[-2]=='Variant'):
                disease_dct[line.split()[3]] = line.split()[-1]
```
