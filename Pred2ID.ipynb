```python
def get_taxonomy_id(species_name):
    url = f"https://www.ncbi.nlm.nih.gov/taxonomy/?term={species_name}"
    response = requests.get(url)
    html_text = response.text
    start_index = html_text.find("data-taxid=") + len("data-taxid=")
    end_index = html_text.find(">", start_index)
    taxonomy_id = html_text[start_index:end_index].strip('"')
    return taxonomy_id

# Example usage:
taxonomy_id = get_taxonomy_id("Homo sapiens")
print(taxonomy_id) # prints "9606"
```
